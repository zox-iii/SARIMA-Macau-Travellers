# 【11】REi评价指标绘图

## 0-导入包

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 1-读取数据

```python
## read data 
dfname=['TotalREi','LandREi','AirREi','SeaREi']  # 文件名
allREi = []
for i in range(0,4):
    REi = pd.read_excel('%s.xlsx'%dfname[i],header=None)
    REi.iloc[:,1]=REi.iloc[:,1]
    REi = REi.values[1:,1].astype('float')
    # print(REi)
    allREi.append(REi)
#print(allREi)
```

## 2-绘图

```python
## plot figure
month_list = ['2020.3','2020.4','2020.5','2020.6','2020.7','2020.8','2020.9','2020.10','2020.11','2020.12',
              '2021.1','2021.2','2021.3','2021.4','2021.5','2021.6','2021.7','2021.8','2021.9']
style = ['-o','-.','--','--.']
plt.figure(figsize=(24,12))         # 设置画布大小
for i in range(0,4):
    plt.plot(month_list,allREi[i]*100,style[i],label=dfname[i])

plt.rcParams['font.size'] = '15'
plt.legend()                                          # 添加图例
plt.xlabel("Time(months)")                            # xlabel 
plt.ylabel("REi(%)")                                  # ylabel
plt.title('Recovery ratio of number of passengers to Macau',fontsize = 20)   # 添加标题
plt.savefig('Recovery ratio of number of passengers to Macau.jpg',dpi=256)
plt.show()                                            # 显示图像
plt.close()                                           # 关闭图像
```
