# 【1】2008-2020年的来澳旅客总人数（ByYear）

## 0-导入包

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.tsa.api as smt
import statsmodels.tsa.arima.model as smt2
from statsmodels.stats.diagnostic import acorr_ljungbox as lb_test
```

## 1-时序图

```python
## read data
dfname='2008-2020ByYear'
y=pd.read_csv('%s.csv'%dfname,header=None,encoding='gbk')
y.iloc[:,1]=y.iloc[:,1].astype('float')
y=y.values[:,1]

## plot figure
plt.figure(figsize=(18,9))
plt.plot([i for i in range(2008,2021)],y,'*-')
plt.rcParams['font.size'] = '20'
plt.xlabel("Time(Year)")                            # xlabel 
plt.ylabel("Number of total passengers(person)")                        # ylabel
plt.title('Sequence chart By Year of total from 2008-2020')        # 添加标题
plt.savefig("SequenceChart By Year of total from 2008-2020.jpg",dpi=256)
plt.show()
plt.close()
```

【结论】
1. 2008-2019整体呈上升趋势
2. 2020年由于疫情来澳旅客人数急剧下降
