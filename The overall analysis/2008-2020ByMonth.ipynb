# 【2】2008-2020年的来澳旅客总人数（ByMonth）

## 0-导入包

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.tsa.api as smt
import statsmodels.tsa.arima.model as smt2
from statsmodels.stats.diagnostic import acorr_ljungbox as lb_test
```

## 1-时序图

```python
## read data
dfname='按入境渠道统计之来澳旅客总人次'
y=pd.read_excel('%s.xlsx'%dfname,header=None)
y.iloc[:,1]=y.iloc[:,1].astype('float')
y=y.values[:,1]

## plot figure
plt.figure(figsize=(20,10))
plt.rcParams['font.size'] = '15'
month_list=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

for year in range(2008,2021):
    plt.plot(month_list,y[(year-2008)*12:(year-2008+1)*12],'o-',label='%s'%year)

plt.legend(loc="upper right")
plt.rcParams['font.size'] = '20'
plt.xlabel("Time(months)")                            # xlabel 
plt.ylabel("Number of total passengers(person)")                        # ylabel
plt.title("SequenceChart By Month of total from 2008-2020")        # 添加标题
plt.savefig("SequenceChart By Month of total from 2008-2020.jpg",dpi=256)
plt.show()
plt.close()
```
